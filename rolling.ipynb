```python
def Dstathist(hist_values, **kwargs):
    histogram = np.histogram(hist_values, bins=np.logspace(-2,0,100))
    bincenters = 0.5*(histogram[1][1:]+histogram[1][:-1])
    plt.plot(bincenters, histogram[0], **kwargs)
    
```
