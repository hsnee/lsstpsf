```python
def getCoadds(obj):
    return (obj['metricValues'])[obj['mask']==False]
```

```python
plt.hist(getCoadds(pontus_2489_CoaddM5), bins=200, histtype='step', linewidth=2, label='pontus_2489');
plt.hist(getCoadds(kraken_2035_CoaddM5), bins=200, histtype='step', linewidth=2, label='kraken_2035');
plt.hist(getCoadds(pontus_2502_CoaddM5), bins=200, histtype='step', linewidth=2, label='pontus_2502');
plt.legend()
plt.title('coadded depth histogram')
```

```python
def getBins(histmetric):
    a = histmetric['slicePoints']
    return ((a.flatten()[0])['bins'])[1:]
```

```python
kraken_2035_airmass = np.load('kraken_2035_airmass_Histogram_WFD_r_band_ONED.npz')
kraken_2035_airmass_bins = getBins(kraken_2035_airmass)
pontus_2489_airmass = np.load('pontus_2489_airmass_Histogram_WFD_r_band_ONED.npz')
pontus_2489_airmass_bins = getBins(pontus_2489_airmass)
pontus_2502_airmass = np.load('pontus_2502_airmass_Histogram_WFD_r_band_ONED.npz')
pontus_2502_airmass_bins = getBins(pontus_2502_airmass)
```

```python
plt.plot(kraken_2035_airmass_bins, kraken_2035_airmass['metricValues'], label='kraken_2035')
plt.plot(pontus_2489_airmass_bins, pontus_2489_airmass['metricValues'], label='pontus_2489')
plt.plot(pontus_2502_airmass_bins, pontus_2502_airmass['metricValues'], label='pontus_2502')
plt.legend()
plt.title('airmass histogram')
```
